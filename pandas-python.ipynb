Pandas id a powerful data manipulation libraries in python,widely used for data analysis and cleaning . It provides two primary data structure. Series and DataFrame. A series is a one-dimensional array-like object, while a DataFrame is two dimensional ,size-mutable , and potentially hetrogeneous tabular data structure with ;abled axes(rows and columns)

1. **Pandas series:**

Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.).

```python
#Series
data = [1,2,3,4,5]
series =pd.Series(data)
print(series)
print(type(series))
```

```python
#create Series from dictionary
data= {'a':1,'b':2,'c':3}
series_dic =pd.Series(data)
print(series_dic)
print(type(series_dic))
```

```python
data = [10,20,30]
index =['a','b','c']
print(pd.Series(data = data,index = index))
print(type(pd.Series(data = data,index = index)))
```

2. **DataFrame:**

 *  Syntex: *class pandas.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)*
  *  Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.

```python
#Creating dataframe from a dictionary of list
data= {
    'Name':['rajesh','john','jack'],
    'Age':[25,30,45],
    'City':['Banglore','NewYork','florida']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
#data Aggregating and grouping 
df.head()
grouped_mean = df.groupby('COUNTRY')['SALES'].mean()
print(grouped_mean)
```

```python
#multiple aggregate function 
grouped_agg = df.groupby('COUNTRY')['SALES'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
#merging and joining of dataframe

# Example dataframes
df1 = pd.DataFrame({'Key': [1, 2, 3], 'Value1': ['A', 'B', 'C']})
df2 = pd.DataFrame({'Key': [2, 3, 4], 'Value2': ['X', 'Y', 'Z']})
print(df1)
print(df2)
# Merging using inner join
merge_df = pd.merge(df1, df2, on="Key", how="inner")
print(merge_df)
```

```python

# Merging Outer join
merge_df = pd.merge(df1, df2, on="Key", how="outer")
print(merge_df)
```

```python
# Merging left join
merge_df = pd.merge(df1, df2, on="Key", how="left")
print(merge_df)
```

```python
# Merging right join
merge_df = pd.merge(df1, df2, on="Key", how="right")
print(merge_df)
```
