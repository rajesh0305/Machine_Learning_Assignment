SQL And SQLite

SQL(structured query language) is a standard language for managing and manipulating relational database.

SQLite is a self-contained ,serverless ,and zero-configuration database engine that is widely used for embedded database system.

```python
import sqlite3 
```

```python
#connect to an SQLite database
connection =sqlite3.connect('example.db')
connection 
```

```python
cursor = connection.cursor()
```

```python
#create a table
cursor.execute('''
   Create Table If Not Exists employees(
       id Inyeger Primary Key,
       name Text Not Null,
       age Integer,
       department text
   )
''')
#commit the cahnges
connection.commit()
```

```python
cursor.execute('''
   Select * from employees
''')
```

```python
#Insert tha data into SQLite databse table
cursor.execute('''
Insert Into employees (name,age,department)
       values('rajesh',25,'Data Scientist')
''')

cursor.execute('''
Insert Into employees (name,age,department)
       values('manav',24,'Data Engineer')
''')
cursor.execute('''
Insert Into employees (name,age,department)
       values('anuj',27,'Analog Engineer')
''')
#connection.commit()
connection.commit()
```

```python
#query data from the table 
cursor.execute('select * from employees')
rows = cursor.fetchall()

#print the query data
for row in rows:
    print(row)
    
```

```python
#Update the data into tha table
cursor.execute('''
  UPDATE employees
  Set age = 27
  where name = 'anuj';
''')
connection.commit()
```

```python
#query data from the table 
cursor.execute('select * from employees')
rows = cursor.fetchall()

#print the query data
for row in rows:
    print(row)
```

```python
#Delete the data from the table 
cursor.execute('''
Delete from employees
    where name ='anuj'
''')
connection.commit()
```

```python
#query data from the table 
cursor.execute('select * from employees')
rows = cursor.fetchall()

#print the query data
for row in rows:
    print(row)
```

```python
#working with sales data
connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

#create a table for sales data
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales(
   id Integer Primary Key,
   date text not null,
   product text not null,
   sales Integer,
   region text
   )
''')
#insert some of data
sales_data =[
   ('2023-01-01','p1',100,'north'),
    ('2023-01-02','p2',200,'west'),
    ('2023-01-03','p3',500,'south'),
    ('2023-01-04','p4',400,'east'),
    ('2023-01-05','p5',300,'north')
]

cursor.executemany('''
Insert into sales(date,product,sales,region)
      values(?,?,?,?)
''',sales_data)
connection.commit()

```

```python
#query data from the table 
cursor.execute('select * from sales')
rows = cursor.fetchall()

#print the query data
for row in rows:
    print(row)
```

```python
#close the connection parmanently
connection.close()
```
