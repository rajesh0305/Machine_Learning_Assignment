Why we are doing the Column/Row normalization in the dataset?

ans:-we are doing normalization in the dataset for the **scaling** the data in the dataset.

**Lets Create some Dummy dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame({'Height(cm)':[162,172,182,160,155,168,170,190,173.9,166.5],
                   'Weight(Kg)':[56,72,84,58,67,80,56.8,90,67,88]})
df
```

```python
print(df.describe())
```

```python
plt.scatter(df['Height(cm)'],df['Weight(Kg)'])
plt.xlabel('Height(cm)')
plt.ylabel('Weight(Kg)')
plt.title('Height vs Weight')
plt.show()
```

Analysis:You can not say any thing about the relation between the "height" and the "weight" because of the scalling of the height and the weight in two different parameter like Kg and cm. so,we must have to scale the dataset.Either we can do by colum normalization or the row normalization.


Using **Column Normalization**

formula for the column normalization is look like:-(ith datapoint-min(column))/(max(column)-min(column))

```python
min1 =df['Height(cm)'].min()
max1 =df['Height(cm)'].max()

min2 =df['Weight(Kg)'].min()
max2 =df['Weight(Kg)'].max()

print(min1,max1)
print(min2,max2)

for i in range(0,len(df['Height(cm)'])):
  df['Height(cm)'][i] = (df['Height(cm)'][i]-min1)/(max1-min1)
  df['Weight(Kg)'][i] = (df['Weight(Kg)'][i]-min2)/(max2-min2)

df

```

**Analysis**: We can see that , After Column normalization the column are scaled ,the data in the column present within the range of (0 and 1).

```python
mean1 = df['Height(cm)'].mean()
mean2 = df['Weight(Kg)'].mean()


plt.scatter(df['Height(cm)'],df['Weight(Kg)'],color = 'red',marker = 'x')
plt.xlabel('height(cn)')
plt.ylabel('Weight(kg)')
plt.title('Height vs Weight')
plt.show()
```

```python
print(df.describe())
```

```python
plt.scatter(df['Height(cm)'], df['Weight(Kg)'], color='blue', label='Data Points')  # Plot the scatter plot of the data

# Plot the mean as a single point
plt.scatter(mean1, mean2, color='red', marker='x', s=100, label='Mean')

plt.xlabel('height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height vs Weight')
plt.legend() # Show the legend to distinguish between data points and the mean
plt.show()
```
