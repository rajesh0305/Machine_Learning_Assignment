Why we are doing **Column Standardization** During datapreprocessing?

Ans: It is performed during the data pre-processing to handle highly varying magnitudes or values or units. If feature scaling is not done, then a machine learning algorithm tends to weigh greater values, higher and consider smaller values as the lower values, regardless of the unit of the values.

Column standardization is also called Mean Centric followed by scaling.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= np.random.randint(100,200,size = (100,2))
df = pd.DataFrame(data,columns=['height','weight'])
df.head()
```

```python
print(df.describe())
```

```python
plt.scatter(df['height'],df['weight'])
plt.xlabel('height')
plt.ylabel('weight')
plt.title('height vs weight')
plt.show()
```

Start Doing **Column Standardization**

formula:  A' = (A(i)-A(bar))/(S)

Where:

 A(i):-datapoint at ith location

A(bar):-Mean of the column

S:-Standard deviation of the column

**Condition:** overall mean must be zero and Standard deviation must be one.

```python
mean1 = df['height'].mean()
std1 = df['height'].std()

mean2 = df['weight'].mean()
std2 = df['weight'].std()

for i in range (len(df)):
  df['height'][i] = (df['height'][i]-mean1)/std1
  df['weight'][i] = (df['weight'][i]-mean2)/std2

df.head()
```

```python
print(df.describe())
```

```python
plt.scatter(df['height'],df['weight'],color= 'green',marker = 'x',label ='DataPoint')
plt.scatter(std1,std2,label ='Std-Dev')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('height vs weight')
plt.legend()
plt.show()
```

```python
plt.scatter(df['height'],df['weight'],color= 'green',marker = 'x',label ='DataPoint')
plt.scatter(mean1,mean2, color='red', marker='x', label='Mean')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('height vs weight')
plt.legend()
plt.show()
```

```python
plt.scatter(df['height'],df['weight'],color= 'green',marker = 'x',label ='DataPoint')
plt.scatter(df['height'].mean(),df['weight'].mean(), color='red', marker='x', label='Mean')
# plt.scatter(std1,std2,label ='Std-Dev')
plt.scatter(df['height'].std(),df['weight'].std(),label= 'Std_Dev')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('height vs weight')
plt.legend()
plt.show()
```
