# Target Guided Ordinal Encoding 

it is a techniques used to encode categorical variables based on their relationship with the target variable.This encoding techniques is useful when we have a categorical variable with a large number of unique categories,and we want to use this variale as a feature in our machine learning model.

In target guided ordinal encoding ,we replace each category in teh categories variable with a numerical value based on the mean or median of the target variable for that categories .This creates a monotonic relationship between the categoical variable and the target variavle ,which can imprve the predictive power of our model.


```python
import pandas as pd 
```

```python
#creating a simple dataframe with a categorical variav=ble and the target 
df =pd.DataFrame({
    'city':['new york','london','paris','tokyo','new york','paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
df
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
df['city_encoder'] = df['city'].map(mean_price)
```
