```python
#error/single key is always used in dictionaries 
student = {"name":"rajesh","age":24,"name":46}
print(student)
```

```python
#Accessing dictionaries from the element 
student = {"name":"rajesh","age":24,"grade":'A'}
print(student)
print(student['grade'])
print(student['age'])
print(student['name'])

#accessing using get() method 
print("accessing using get() method ")
print(student.get('grade'))
print(student.get('name'))
# print(student.get('last_neme'))
print(student.get('last_name',"Not Available"))
```

```python
#modifying the dictionaries element
#Dictionaries are mutable ,so you can add, update or delete element
print(student)
print("updating the value of key")
student["age"] =33

print(student)

print("adding the new data(key and value)")
student["address"] = "India"

print(student)

#deleting the 
print("deleting the key ")
del student["grade"]
print(student)
```

```python
#dictionaries method 
keys = student.keys()
print(keys)
values = student.values()
print(values)

items =student.items()
print(items)
```

```python
#sshallow copy
student_copy =student
print(student)
print(student_copy)

student["name"] = "manshaa"
print(student)
print(student_copy)

student_copy1 = student.copy()
print(student_copy1)
print(student)


student["name"] ="manshaa1"
print(student_copy1)
print(student)
```

```python
#iterating over dictionaries 
#you can use loops to iterate over dictionaries ,keys ,values ,items

#iterating over keys
for keys in  student.keys():
    print(keys)
```

```python
#iterating over values 
for values in student.values():
    print(values)
```

```python
#iterating over key,value pairs
for key,value in student.items():
    print(f"{key}:{values}")
    # print("key",key)
    # print("value",value)
```

```python
#nested dictionaries 
students ={
    "student1":{"name":"rajesh","age":24,"grade":'A'},
    "student2":{"name":"peter","age":23,"grade":'B'}
    
}
print(students)
#accessing the nested dictionaries 
print("accessing the nested dictionaries ")
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
#iterating ove nested dictionaries 
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
    
```

```python
#dictionaries comphrehension
squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
#condition dictionaries comphrehension 
even = {x:x**2 for x in range(10) if x%2==0}
print(even)
```

```python
#example
#use a dictionaries to count the frequency of the element in the list 
num = [1,2,1,2,1,3,4,5,3,2,4,3,2,4,5]
frq = {}
for num in num:
    if num in frq:
        frq[num]+= 1
    else:
        frq[num] = 1

print(frq)
        
        
```

```python
#merging the two dictionaries 
dist1 = {"a":1,"b":2}
dist2 = {"b":3,"c":4}
merge_dist ={**dist1,**dist2}
print(merge_dist)
```
