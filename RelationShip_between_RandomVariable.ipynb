implementation of the SPEARMAN Rank Correlation Coefficient

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import spearmanr
from scipy.stats import pearsonr
from scipy.stats import linregress

x1 =np.array([7.1,7.1,7.2,8.3,9.4,10.5,11.5])
x2 = np.array([2.8,2.9,2.8,2.6,3.5,4.6,5.6])

plt.scatter(x1,x2)
plt.title('Spearman Rank Correlation Coefficient')
plt.xlabel('-:first random variable:-')
plt.ylabel('-:second random variable:-')
plt.show()

res =spearmanr(x1,x2)
print(res)
print(res.statistic)
print(res.pvalue)
```

```python
plt.scatter(x1,x2)
plt.title('Pearson Correlation Coefficient')
plt.xlabel('-:first random variable:-')
plt.ylabel('-:second random variable:-')
plt.show()

res =pearsonr(x1,x2)
print(res)
print(res.statistic)
print(res.pvalue)
```
