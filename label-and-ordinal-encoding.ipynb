# Level Encoding 

level encoding and ordinal encoding are two techniques used to encode categorical data as numerical.


Label encoding involves assigning numerical label to each category in the variable.The labels are usually assigned in alphabetical order or based on the frequency of the categories.For example,if we have a categorical variable "color" with three possible values(red,green,blue,we cam represent it using label encoding as follows:
1. Red:1
2. Green:2
3. Blue:3


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    'color':['red','blue','green','green','red']
})
df
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder =LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
print("level encoding for red color:",lbl_encoder.transform([['red']]))
print("level encoding for green color:",lbl_encoder.transform([['green']]))
print("level encoding for blue color:",lbl_encoder.transform([['blue']]))
```

**Ordinal Encoding** 

it is used to encode categorical data that have an intrinsic order or ranking .In this technique,each category is assigned a numerical value based on its position in the order ,for example ,if we have a categorical variable "education level" with four possible values (high school,collage,graduate,post_graduate),we can represent it using ordinal encoding as follows:
1. High School:1
2. College:2
3. Garduate:3
4. Pst Graduate:4

```python
#Oridinal Encoding 
from sklearn.preprocessing import OrdinalEncoder
```

```python
#creating simple dataframe
df = pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
df
```

```python
#create an instance of ordinal encoding and then fit_transform
encoder = OrdinalEncoder(categories = [['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```
