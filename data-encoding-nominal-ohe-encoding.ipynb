# Data Encoding

As we all know that better encoding leads to a better model and most algorithms cannot handle the categorical variables unless they are converted into a numerical value.
1. Nominal/OHE Encoding
2. Label and Ordinal Encoding
3. Target Guided Ordinal Encoding 

1. **Nominal/One Hot Encoding**

**One Hot Encoding** is a method for converting categorical variables into a binary format. It creates new binary columns (0s and 1s) for each category in the original variable. Each category in the original column is represented as a separate column, where a value of 1 indicates the presence of that category, and 0 indicates its absence.

One hot encoding ,also known for as nominal encoding ,is a technique used to represent categorical data as numerical data,which is more suitable for the machine learning algorithm. In this techniques ,each category is represented as binary vector where each bit correspond to a unique category .for examole ,if we have categorical variable "color" with three possible values(red,green,blue),we can represent it using one hot encoding like:
1. Red:[1,0,0]
2. Green:[0,1,0]
3. Blue:[0,0,1]

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
#creating a simple dataframe 
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
#create an instance of onehotencoder
encoder = OneHotEncoder()
```

```python
#perform fit() and Transform
result= encoder.fit_transform(df[['color']])
print(type(result))
print(result.toarray())
```

```python
import pandas as pd
encoder_df = pd.DataFrame(result.toarray(),columns = encoder.get_feature_names_out())
encoder_df
```

```python
pd.concat([df,encoder_df],axis = 1)
```
