```python
#while loop
## the while loop continue to execute as long as the condition id True
count =0
while count<6:
    print(count)
    count = count+1
    
```

```python
count = 10
while count%2==0:
    print(count)
    count =count+1
```

```python
#loop control statement
## break
## the break stmt exits the loop permanently

for i in range(10):
    if i==5:
        break
    print(i)  
#continue
print('/n')
#the continue stmt skip the current iteration and continue with the text
for i in range(10):
    if i%2==0:
        continue
    print(i)

#pass
#the pass stmt is a null operation ; it does nothing 
print("\n")
for i in range(10):
    if i==3:
        pass
    print(i)    
```

```python
##nested loops 
## a loop inside a loops 
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## example calcu;ate the sum of first n natural number using a while loop
## while loop
n = 10
sum =0
count= 1
while count<=n:
    sum = sum+count
    count = count+1

print("sum of first n natural number ",sum)
```

```python
#find prime number btwn 1 to 100
for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
            else:
                print(num)
```
