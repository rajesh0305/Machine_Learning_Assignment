```python
import seaborn as sns
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Display the first few rows
iris.head()
```

# Choosing the Right Plot

1. **Univariate Analysis**: Use histplot, kdeplot, or boxplot.
2. **Bivariate Analysis**: Use scatterplot, lineplot, or heatmap.
3. **Categorical Data**: Use barplot, violinplot, or boxplot.
4. **Multi-variable Analysis**: Use pairplot or heatmap

**1. Distributions**
   
Used to understand the distribution of data.

Histogram: sns.histplot

```python
sns.set(style = "darkgrid")
sns.histplot(data=iris, x="sepal_length", kde=True)#kde=kernal density estimation
```

```python
#KDE Plot: sns.kdeplot
#(Kernel Density Estimate)
sns.kdeplot(data=iris, x="sepal_length", hue="species", fill=True)
```

3. ECDF Plot:
   sns.ecdfplot
(Empirical Cumulative Distribution Function)

```python
sns.ecdfplot(data=iris, x="sepal_length", hue="species")
```

```python
x = iris.drop(columns=['species'])
y = iris['species']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
print("x_tarin ,x_test",x_train.shape,x_test.shape)
print("y_train ,y_test",y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data
scaler.fit(x_train)

# Transform the training and testing data
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier  = KNeighborsClassifier(n_neighbors = 5,leaf_size=20)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score

# Assuming y_test and y_pred are defined
accuracy = accuracy_score(y_test, y_pred)

# Convert to percentage
accuracy_percentage = accuracy * 100

print(f"Accuracy: {accuracy_percentage:.2f}%")
```
