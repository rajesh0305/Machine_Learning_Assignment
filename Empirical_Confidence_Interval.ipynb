```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.utils import resample
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
```

```python
data =np.array([180,162,158,172,168,150,171,183,165,175])

n_iteration = 1000
n_size = int(len(data))

median = list()
for i in range(n_iteration):
  s = resample(data,n_samples=n_size)
  m =np.median(s)
  median.append(m)

plt.hist(median)
plt.title("Empirical confidence interval")
plt.xlabel("data")
plt.ylabel("n_iteration")
plt.show()

#now we calculating the Empirical confidence interval
alpha = 0.95
p = ((1-alpha)/2)*100
lower  = np.percentile(median,p)

p = (alpha+(1-alpha)*100)
upper = np.percentile(median,p)

print("lower",lower)
print("upper",upper)
```
